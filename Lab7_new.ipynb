```python
import pandas as pd
import numpy as np
```

```python
def create_file(p1):
    myvar = np.random.choice([0, 1], size=838860800, replace=True, p=[p1, 1-p1])
    myvar = np.packbits(myvar)
    return myvar

```

```python
myvar_50 = create_file(0.5)
myvar_60= create_file(0.6)
myvar_70= create_file(0.7)
myvar_80= create_file(0.8)
myvar_90= create_file(0.9)
myvar_100= create_file(1.0)

```

```python
open('zeros_100p', 'wb').write(myvar_50)
open('zeros_90p', 'wb').write(myvar_60)
open('zeros_80p', 'wb').write(myvar_70)
open('zeros_70p', 'wb').write(myvar_80)
open('zeros_60p', 'wb').write(myvar_90)
open('zeros_50p', 'wb').write(myvar_100)
```
