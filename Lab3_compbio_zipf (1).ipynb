```python
with open ("romeoandjuliet.txt", "r") as myfile:
    data=myfile.readlines()
```

```python
#print(data)
```

```python
import string
for i in range(len(data)):
    data[i] = data[i].replace('\n', ' ')
    data[i] = data[i].replace('.', ' ')
    data[i] = data[i].replace(',', ' ')
    data[i] = data[i].replace('?', ' ')
    data[i] = data[i].replace(';', ' ')
    data[i] = data[i].replace('-', ' ')
    data[i] = data[i].replace('\'', ' ')
    data[i] = data[i].replace('(', ' ')
    data[i] = data[i].replace(')', ' ')
    data[i] = data[i].replace('[', ' ')
    data[i] = data[i].replace(']', ' ')
    data[i] = data[i].replace('"', ' ')
    data[i] = data[i].replace('"', ' ')
    data[i] = data[i].replace('!', ' ')
    data[i] = data[i].replace(':', ' ')
```

```python
data_without = ""
for i in data:
    data_without = data_without+i
```

```python
o = data_without.split(" ")
```

```python
x_lower = [element.lower() for element in o] 
```

```python
#print(x_lower)
```

```python
dictionary=[]
for i in x_lower:
    if i == '':
        continue
    else:
        dictionary.append(i)
```

```python
#print(dictionary)
```

```python
diction=set(dictionary)
```

```python
#print(diction)
```

```python
Y = list(diction)
```

```python
#print(Y)
```

```python
dictionary_1=[]
Y.sort()
for i in sorted(Y):
    dictionary_1.append(i)
```

```python
#print(dictionary_1)
```

```python
count =[]
for i in dictionary_1:
    count1=0
    for j in dictionary:
        if i==j:
            count1=count1+1
        else:
            continue
    count.append(count1)
```

```python
#print(count)
```

```python
len(count)
```

```python
len(dictionary_1)
```

```python
dictionary_freq = dict(zip(dictionary_1, count))
#print(dictionary_freq)
```

```python
x = [i for i in range(len(count))]
```

```python
y = [j for j in dictionary_freq.values()]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(x,y)
my_xticks = list(dictionary_freq.keys())
plt.xticks(x, my_xticks)
fig = plt.gcf()
fig.set_size_inches(50,32)
```
